```python

#Frequency distribution and mode
#frequency=the number of rows for each element for a specific feature
#mode=the most popular element of a specific feature


def print_distr(cab_data, column_key):
    print(column_key.upper())

    counts = Counter()
    for row in cab_data:
        counts[row[column_key]] += 1

    for k,v in sorted(counts.items()):
        print('* {}: {}'.format(k, v))




print_distr(bank_data,'job')
print_distr(bank_data,'marital')
print_distr(bank_data,'education')
print_distr(bank_data,'default')
print_distr(bank_data,'housing')
print_distr(bank_data,'loan')
print_distr(bank_data,'contact')
print_distr(bank_data,'month')
print_distr(bank_data,'poutcome')
print_distr(bank_data,'y')

print_distr(bank_additional_data,'job')
print_distr(bank_additional_data,'marital')
print_distr(bank_additional_data,'education')
print_distr(bank_additional_data,'default')
print_distr(bank_additional_data,'housing')
print_distr(bank_additional_data,'loan')
print_distr(bank_additional_data,'contact')
print_distr(bank_additional_data,'month')
print_distr(bank_additional_data,'day_of_week')
print_distr(bank_additional_data,'pdays')
print_distr(bank_additional_data,'previous')
print_distr(bank_additional_data,'poutcome')
print_distr(bank_additional_data,'emp.var.rate')
print_distr(bank_additional_data,'nr.employed')
print_distr(bank_additional_data,'y')



```

```python
# defines a new 'mode' function that takes two parameters
def mode(data, column_key):
    mode_counter = Counter()
    for row in data:
        mode_counter[row[column_key]] += 1
    return mode_counter.most_common(1)[0][0]

# example on how to use the 'mode' function


#stats(bank_data,'y')


#print("TRANSACTION_ID mode:", mode(bank_data,'y'))
#print("DATE_OF_TRAVEL mode:", mode(cab_data,DATE_OF_TRAVEL))
#print("COMPANY mode:", mode(cab_data,COMPANY))
#print("CITY mode:", mode(cab_data,CITY))


```
